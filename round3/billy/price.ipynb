```python
import pandas as pd
import matplotlib.pyplot as plt

# correlation
df0 = pd.read_csv("../data/prices_round_3_day_0.csv", sep=";")
df1 = pd.read_csv("../data/prices_round_3_day_1.csv", sep=";")
df2 = pd.read_csv("../data/prices_round_3_day_2.csv", sep=";")

df0 = df0[df0["product"] == "BERRIES"]
df1 = df1[df1["product"] == "BERRIES"]
df2 = df2[df2["product"] == "BERRIES"]

df = pd.concat([df0, df1, df2], ignore_index=True)
```

```python
plt.plot(df["mid_price"])
plt.show()
# lowest: 20000 / 100 = 200
```
